```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Assuming df is your DataFrame containing 'longitude' and 'latitude' columns

# Selecting features for clustering
geographical_data = df[['longitude', 'latitude']]

# Standardize the features (important for KMeans)
scaler = StandardScaler()
geographical_data_scaled = scaler.fit_transform(geographical_data)

# Apply KMeans clustering
n_clusters = 6  # You can adjust the number of clusters based on your analysis
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(geographical_data_scaled)

# Calculate cluster centroids
cluster_centroids = scaler.inverse_transform(kmeans.cluster_centers_)

# Function to determine cardinal direction based on coordinates
def get_cardinal_direction(x, y):
    angle = (180 / np.pi) * np.arctan2(y, x)
    if angle < 0:
        angle += 360
    if 22.5 <= angle < 67.5:
        return 'NE'  # Northeast
    elif 67.5 <= angle < 112.5:
        return 'E'   # East
    elif 112.5 <= angle < 157.5:
        return 'SE'  # Southeast
    elif 157.5 <= angle < 202.5:
        return 'S'   # South
    elif 202.5 <= angle < 247.5:
        return 'SW'  # Southwest
    elif 247.5 <= angle < 292.5:
        return 'W'   # West
    elif 292.5 <= angle < 337.5:
        return 'NW'  # Northwest
    else:
        return 'N'   # North

# Assign cardinal direction labels to clusters
df['direction'] = df.apply(lambda row: get_cardinal_direction(row['longitude'] - cluster_centroids[row['cluster']][0],
                                                              row['latitude'] - cluster_centroids[row['cluster']][1]),
                           axis=1)

# Visualize the clusters on a scatter plot with direction labels
plt.figure(figsize=(12, 10))
sns.scatterplot(x='longitude', y='latitude', hue='cluster', data=df, palette='viridis', s=100)
for i in range(n_clusters):
    plt.text(cluster_centroids[i][0], cluster_centroids[i][1], f'Cluster {i + 1} ({df["direction"].loc[df["cluster"] == i].iloc[0]})',
             fontsize=12, bbox=dict(facecolor='white', alpha=0.7))
plt.title('Geographical Clustering of Police Killings with Cardinal Direction Labels')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Selecting features for clustering
geographical_data = df[['longitude', 'latitude']]

# Standardize the features (important for KMeans)
scaler = StandardScaler()
geographical_data_scaled = scaler.fit_transform(geographical_data)

# Apply KMeans clustering
n_clusters = 7  # You can adjust the number of clusters based on your analysis
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(geographical_data_scaled)

# Calculate the density of each cluster
cluster_density = df['cluster'].value_counts(normalize=True).to_dict()

# Map cluster densities to labels
density_labels = {cluster: 'High' if density > 0.25 else 'Low' if density < 0.1 else 'Medium' for cluster, density in cluster_density.items()}

# Create a new column 'density_label' in the DataFrame
df['density_label'] = df['cluster'].map(density_labels)

# Visualize the clusters with density labels on a scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='longitude', y='latitude', hue='density_label', data=df, palette='viridis', s=100)
plt.title('Geographical Clustering of Police Killings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
